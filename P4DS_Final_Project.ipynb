<h3 id="information">1.1. Thông tin bộ dữ liệu</h3>

##### **Chủ đề chính của bộ dữ liệu**
Bộ dữ liệu tập trung vào các yếu tố nguy cơ liên quan đến bệnh tim, bao gồm thông tin về tình trạng sức khỏe, lối sống và thói quen cá nhân của người tham gia khảo sát. Các biến số như huyết áp cao, mức cholesterol, tình trạng tiểu đường, chỉ số BMI, thói quen hút thuốc và sử dụng rượu bia đều được ghi nhận. Mục tiêu chính của bộ dữ liệu là cung cấp thông tin giúp phân tích và phát hiện các yếu tố có thể dự báo nguy cơ mắc bệnh tim, nhằm đưa ra biện pháp phòng ngừa kịp thời.

##### **Phương pháp thu thập dữ liệu**
Dữ liệu được thu thập từ CDC thông qua các cuộc khảo sát điện thoại trong hệ thống BRFSS ( Behavioral Risk Factor Surveillance System), bao gồm các thông tin sức khỏe cá nhân từ năm 2020 đến 2022. Bộ dữ liệu gốc đã được chuyển đổi từ định dạng SAS sang CSV. Tác giả đã sử dụng Python với thư viện pandas để tổ chức và làm sạch dữ liệu, sau đó chọn lọc các biến số có ảnh hưởng trực tiếp hoặc gián tiếp đến bệnh tim mạch. Có thể xem chi tiết về quá trình tiền xử lý dữ liệu tại đây: [data_processing.ipynb](https://github.com/kamilpytlak/data-science-projects/blob/main/heart-disease-prediction/2022/notebooks/data_processing.ipynb).

##### **Sử dụng dữ liệu và tính hợp pháp**
Bộ dữ liệu này được trích xuất từ CDC và là một phần của hệ thống giám sát các yếu tố nguy cơ hành vi liên quan đến sức khỏe (BRFSS), do đó hoàn toàn hợp pháp khi sử dụng cho mục đích nghiên cứu và giáo dục, nhóm có thể sử dụng nó để phân tích và khám phá các yếu tố nguy cơ liên quan đến bệnh tim nhằm tìm ra cách dự báo và phòng ngừa bệnh.

Sử dụng cá nhân: nhóm có quyền truy cập và sử dụng dữ liệu cho mục đích cá nhân và giáo dục như đồ án môn học và không liên quan tới mục đích thương mại.

#### 2.4.2. Có dòng dữ liệu nào bị trùng không?
Sử dụng Pandas `duplicated()` để kiểm tra xem có bất kỳ dòng trùng lặp nào không.  
- Khai báo biến `duplicated_rows` để lưu trữ số lượng các dòng trùng lặp.  
- Nếu `duplicated_rows` > 0:  
    - In ra các dòng trùng lặp này và sau đó loại bỏ chúng.  
    - Kiểm tra: Nếu `duplicated_rows` == 0 (False), thì không có dòng nào bị trùng lặp.  
- Nếu không, tiếp tục sang phần tiếp theo.  


```python
duplicated_rows = data_heart.duplicated().sum()
print("Số dòng bị trùng lắp:", duplicated_rows)
```

```python
def plot_bar_chart(df):
    # Lấy các cột dạng categorical
    category_cols = [col for col in df.columns if df[col].dtype == 'O']
    num_cols = len(category_cols)
    ncols = 3  # Số subplot mỗi hàng
    rows = -(-num_cols // ncols)  # Làm tròn lên để tính số hàng

    # Tạo figure
    plt.figure(figsize=(15, 12))
    plt.suptitle(
        'Count plot for various categorical features\n______________________________',
        fontsize=24, fontweight='bold', fontfamily='serif', ha='center', color="#000000"
    )

    counts = 1

    for col in category_cols:
        ax = plt.subplot(rows, ncols, counts)
        sns.countplot(data=df, x=col, hue='HadHeartAttack', width=0.8, ax=ax, palette="coolwarm")
        ax.set_title(col, fontsize=12, fontweight='bold')
        counts += 1

    # Hiển thị biểu đồ
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Giữ khoảng trống cho suptitle
    plt.show()

# Thêm các đặc trưng categorical
data = data_heart[['DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 
                   'DifficultyWalking', 'DifficultyDressingBathing',
                   'DifficultyErrands', 'HadHeartAttack', 'DifficultyConcentrating']].copy()
data['HadHeartAttack'] = data['HadHeartAttack'].apply(lambda x: 1 if x == 'Yes' else 0)

plot_bar_chart(data)
```

#### Phân phối của các đặc trưng phân số

```python
def plot_distplot(df):
    # Lọc các cột kiểu số
    numeric_cols = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]
    num_cols = len(numeric_cols)
    
    # Số cột/hàng trong layout
    ncols = 3  # Số subplot mỗi hàng
    nrows = -(-num_cols // ncols)  # Làm tròn lên để tính số hàng
    
    # Tạo figure và axes
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 5, nrows * 5))
    axes = axes.flatten()  # Chuyển mảng 2D thành 1D để dễ truy cập
    
    # Title chung
    fig.suptitle('Distplot for various numerical features\n_________________', 
                 fontsize=24, fontweight='bold', fontfamily='serif', color="#000000")
    
    for i, col in enumerate(numeric_cols):
        ax = axes[i]
        sns.histplot(df[col], kde=True, color='blue', bins=10, ax=ax)
        ax.set_title(f'Distribution of \'{col}\'', fontsize=12)
        ax.grid(color='#cccccc', linestyle='--', linewidth=0.5)
    
    # Xóa các subplot thừa (nếu số cột không chia hết)
    for j in range(len(numeric_cols), len(axes)):
        fig.delaxes(axes[j])
    
    # Tăng khoảng cách giữa các subplot
    plt.tight_layout(rect=[0, 0, 1, 0.96]) 
    plt.show()

# Ví dụ sử dụng:
plot_distplot(data_heart)
```

#### Phân phối tần suất của các đặc trưng phân số

```python
def plot_single_distribution(df, col, row, col_index):
    """
    Vẽ một cặp subplot cho phân phối của một cột theo biến mục tiêu.
    """
    # Subplot 1: Tiêu đề
    ax_title = plt.subplot(row, 2, col_index)
    ax_title.axis('off')  # Ẩn trục
    ax_title.text(
        0.5, 0.5, 
        f"Distribution of {col}\naccording to\ntarget variable\n___________",
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=18,
        fontweight='bold',
        fontfamily='serif',
        color='#000000'
    )
    
    # Subplot 2: Phân phối dữ liệu
    ax_dist = plt.subplot(row, 2, col_index + 1)
    sns.kdeplot(
        data=df, 
        x=col, 
        hue="HadHeartAttack", 
        fill=True, 
        palette="colorblind", 
        alpha=0.5, 
        linewidth=0.1, 
        ax=ax_dist
    )
    ax_dist.grid(color='#000000', linestyle=':', axis='y', zorder=0, dashes=(1, 5))
    ax_dist.set_title(f"Density of {col}", fontsize=12, fontweight='bold')
    ax_dist.set_xlabel(col, fontsize=10)
    ax_dist.set_ylabel("Density", fontsize=10)
    

def plot_distributions(df, columns):
    """
    Vẽ các phân phối cho một danh sách cột nhất định từ DataFrame.
    """
    num_cols = len(columns)
    plt.figure(figsize=(15, num_cols * 4))  # Điều chỉnh kích thước dựa trên số cột

    for i, col in enumerate(columns):
        plot_single_distribution(df, col, num_cols, i * 2 + 1)
    
    plt.tight_layout()
    plt.show()

# Sử dụng hàm với DataFrame cụ thể
columns_to_plot = ['MentalHealthDays', 'PhysicalHealthDays', 'SleepHours']
plot_distributions(data_heart, columns_to_plot)
```

```python
df = data_heart[['HadStroke','PhysicalActivities','AgeCategory','BMI','SleepHours','SmokerStatus', 'AlcoholDrinkers', 'ECigaretteUsage']].copy()

#Sleep
sleep_bins = [0, 6, 8, 24]  # Giới hạn nhóm
sleep_labels = ['Short sleep (< 6h)', 'Normal sleep (6 - 8h)', 'Long sleep (> 8h)']  # Nhãn nhóm
df['Sleep_Group'] = pd.cut(df['SleepHours'], bins=sleep_bins, labels=sleep_labels, right=False)

#BMI
bmi_bins = [0, 18.5, 25, 30, 100]  # Giới hạn nhóm
bmi_labels = ['Underweight', 'Normalweight', 'Outweight', 'Obesity']  # Nhãn nhóm
df['BMI_Group'] = pd.cut(df['BMI'], bins=bmi_bins, labels=bmi_labels, right=False)

label_mapping_smoker = {
    'Current smoker - now smokes every day': 'Smoker (daily)',
    'Current smoker - now smokes some days': 'Smoker (occasional)',
    'Former smoker': 'Ex-smoker',
    'Never smoked': 'Never smoked'
}

label_mapping_ecig = {
    'Never used e-cigarettes in my entire life': 'Never used',
    'Use them some days': 'Some days',
    'Use them every day':'Every day',
    'Not at all (right now)':'Not using'
}

# Áp dụng mapping để rút gọn nhãn cho SmokerStatus
df['SmokerStatus'] = df['SmokerStatus'].replace(label_mapping_smoker)
df['ECigaretteUsage'] = df['ECigaretteUsage'].replace(label_mapping_ecig)
```

##### Khó khăn nào bạn đã gặp phải?
Việc đặt ra một câu hỏi mà thật sự mang lại ý nghĩa lớn trong thực tiễn, nghiên cứu và chỉ dùng những thuộc tính có sẵn trong tập dữ liệu có lẽ là điều khó khăn nhất đối với mình. Vì việc tìm hiểu ý nghĩa ẩn sau mỗi thuộc tính và kết nối các thuộc tính với nhau gây mất khá nhiều thời gian.
##### Bạn đã học được những gì?

##### Nếu có thêm thời gian, bạn dự định sẽ làm gì?
